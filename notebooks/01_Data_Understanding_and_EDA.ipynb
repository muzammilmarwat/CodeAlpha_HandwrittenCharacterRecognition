# Data Understanding and EDA

This notebook introduces the MNIST handwritten digit dataset and prepares the analysis plan for later modeling phases.

## Introduction

The goal is to understand image dimensions, class balance, pixel intensity ranges, and representative digit samples before building a CNN model.

## Import Libraries

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from tensorflow.keras.datasets import mnist

sns.set_theme(style="whitegrid")
```

## Load MNIST Dataset

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

## Dataset Shape

```python
print(f"Training images: {x_train.shape}")
print(f"Training labels: {y_train.shape}")
print(f"Test images: {x_test.shape}")
print(f"Test labels: {y_test.shape}")
```

## Class Distribution

```python
class_counts = pd.Series(y_train, name="digit").value_counts().sort_index()

plt.figure(figsize=(8, 4))
sns.barplot(x=class_counts.index, y=class_counts.values, color="#4C78A8")
plt.title("MNIST Training Class Distribution")
plt.xlabel("Digit")
plt.ylabel("Count")
plt.show()
```

## Sample Digit Visualization

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for digit, ax in enumerate(axes.ravel()):
    sample_index = np.where(y_train == digit)[0][0]
    ax.imshow(x_train[sample_index], cmap="gray")
    ax.set_title(f"Digit {digit}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

## Pixel Value Analysis

```python
print(f"Minimum pixel value: {x_train.min()}")
print(f"Maximum pixel value: {x_train.max()}")
print(f"Mean pixel value: {x_train.mean():.2f}")

plt.figure(figsize=(8, 4))
plt.hist(x_train.ravel(), bins=50, color="#F58518")
plt.title("Pixel Value Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```
